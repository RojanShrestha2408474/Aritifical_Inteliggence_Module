**Task-1:** Create a Python program that converts between different units of measurement.

```python
def meters_to_feet(meters):
    """
    Convert meters to feet.

    Parameters:
        meters (float): Value in meters

    Returns:
        float: Value converted to feet
    """
    return meters * 3.28084


def feet_to_meters(feet):
    """
    Convert feet to meters.

    Parameters:
        feet (float): Value in feet

    Returns:
        float: Value converted to meters
    """
    return feet / 3.28084


def kilograms_to_pounds(kg):
    """
    Convert kilograms to pounds.

    Parameters:
        kg (float): Value in kilograms

    Returns:
        float: Value converted to pounds
    """
    return kg * 2.20462


def pounds_to_kilograms(lb):
    """
    Convert pounds to kilograms.

    Parameters:
        lb (float): Value in pounds

    Returns:
        float: Value converted to kilograms
    """
    return lb / 2.20462


def liters_to_gallons(liters):
    """
    Convert liters to gallons.

    Parameters:
        liters (float): Value in liters

    Returns:
        float: Value converted to gallons
    """
    return liters * 0.264172


def gallons_to_liters(gallons):
    """
    Convert gallons to liters.

    Parameters:
        gallons (float): Value in gallons

    Returns:
        float: Value converted to liters
    """
    return gallons / 0.264172


def main():
    """
    Main function to run the unit conversion program.
    """
    try:
        print("Choose conversion type:")
        print("L - Length")
        print("W - Weight")
        print("V - Volume")

        choice = input("Enter your choice (L/W/V): ").upper()

        value = float(input("Enter the value to convert: "))

        if choice == 'L':
            option = input("1: Meters to Feet\n2: Feet to Meters\nChoose (1/2): ")
            if option == '1':
                print("Result:", meters_to_feet(value), "feet")
            elif option == '2':
                print("Result:", feet_to_meters(value), "meters")
            else:
                print("Invalid length option.")

        elif choice == 'W':
            option = input("1: Kilograms to Pounds\n2: Pounds to Kilograms\nChoose (1/2): ")
            if option == '1':
                print("Result:", kilograms_to_pounds(value), "pounds")
            elif option == '2':
                print("Result:", pounds_to_kilograms(value), "kilograms")
            else:
                print("Invalid weight option.")

        elif choice == 'V':
            option = input("1: Liters to Gallons\n2: Gallons to Liters\nChoose (1/2): ")
            if option == '1':
                print("Result:", liters_to_gallons(value), "gallons")
            elif option == '2':
                print("Result:", gallons_to_liters(value), "liters")
            else:
                print("Invalid volume option.")

        else:
            print("Unsupported conversion type.")

    except ValueError:
        print("Error: Please enter a numeric value.")


# Run the program
main()
```

**Task-2:** Create a Python program that performs various mathematical operations on a list of numbers

```python
def sum_of_numbers(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
        numbers (list): List of integers or floats

    Returns:
        float: Sum of the numbers
    """
    total = 0
    for num in numbers:
        total += num
    return total


def average_of_numbers(numbers):
    """
    Calculate the average of a list of numbers.

    Parameters:
        numbers (list): List of integers or floats

    Returns:
        float: Average of the numbers
    """
    return sum_of_numbers(numbers) / len(numbers)


def maximum_among_all(numbers):
    """
    Find the maximum number in a list.

    Parameters:
        numbers (list): List of integers or floats

    Returns:
        float: Maximum value
    """
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num


def minimum_among_all(numbers):
    """
    Find the minimum number in a list.

    Parameters:
        numbers (list): List of integers or floats

    Returns:
        float: Minimum value
    """
    min_num = numbers[0]
    for num in numbers:
        if num < min_num:
            min_num = num
    return min_num


def main():
    try:
        print("Choose an operation:")
        print("S  - Sum")
        print("A  - Average")
        print("MA - Maximum")
        print("MI - Minimum")

        choice = input("Enter your choice: ").upper()

        numbers = list(map(float, input(
            "Enter numbers separated by spaces: "
        ).split()))

        if not numbers:
            raise ValueError

        if choice == 'S':
            print("Result:", sum_of_numbers(numbers))
        elif choice == 'A':
            print("Result:", average_of_numbers(numbers))
        elif choice == 'MA':
            print("Result:", maximum_among_all(numbers))
        elif choice == 'MI':
            print("Result:", minimum_among_all(numbers))
        else:
            print("Unsupported operation.")

    except ValueError:
        print("Error: Please enter valid numeric values.")


main()
```

```python
def extract_every_other(lst):
    """
    Extract every other element from a list, starting from the first element.

    Parameters:
        lst (list): Input list

    Returns:
        list: New list containing every other element
    """
    new_lst = []
    for i in range(len(lst)):
        if i % 2 == 0:   # start from first element
            new_lst.append(lst[i])
    return new_lst



lst_item=[1,2,3,4,5];
print(extract_every_other(lst_item));
```

**Exercise on Nested List** bold text

```python
def flatten(lst):
  new_lst=[]
  for subitems in lst:
    for item in subitems:
      new_lst.append(item)
  return new_lst

lst_item=[[1,2],[3,4],[5]];
print(flatten(lst_item));
```

```python
def access_nested_element(lst, indices):
    result = lst
    for index in indices:
        result = result[index]
    return result

lst_item=[[1,2,3],[4,5,6],[7,8,9]]
indice=[1,2];
print(access_nested_element(lst_item,indice));
```

**Sum of All Elements in a Nested List**

```python
def sum_nested(lst):
    result = 0
    for item in lst:
        if isinstance(item, list):
            result += sum_nested(item)
        else:
            result += item
    return result


lst_item=[[1,2],[3,[4,5],6]]
print(sum_nested(lst_item))
```

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            cleaned = remove_element(item, elem)
            result.append(cleaned)
        else:
            if item != elem:
                result.append(item)
    return result

lst = [1, [2, 3, [2, 4]], 2, 5]
print(remove_element(lst, 2))
```

```python
def find_max(lst):
    current_max = None

    for item in lst:
        if isinstance(item, list):
            nested_max = find_max(item)
            if current_max is None or nested_max > current_max:
                current_max = nested_max
        else:
            if current_max is None or item > current_max:
                current_max = item

    return current_max

lst = [[1, 2], [3, [4, 5]], 6]
print(find_max(lst))
```

**Count Occurrences of an Element in a Nested List**

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count
    lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

print(count_occurrences(lst, elem))
```

**Flatten a List of Lists of Lists**

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))  # recursion
        else:
            result.append(item)
    return result

lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(lst))
```

 Nested List Average

```python
def average_nested(lst):
    total = 0
    count = 0

    for item in lst:
        if isinstance(item, list):
            sub_total, sub_count = average_nested(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1

    return total, count

lst = [[1, 2], [3, 4], [5, 6]]
print(average_nested(lst))
```

```python
import numpy as np

# Given matrices
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("1. A * A⁻¹:\n", identity)
print("Is close to identity matrix:", np.allclose(identity, np.eye(2)))

# 2. Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("   BA:\n", BA)
print("AB equals BA?", np.array_equal(AB, BA))

# 3. Prove (AB)^T = B^T A^T
AB_T = (np.dot(A, B)).T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T:\n", AB_T)
print("   B^T A^T:\n", BT_AT)
print("Equality holds?", np.array_equal(AB_T, BT_AT))

# -------------------------------------------------
# Solve system of linear equations using inverse
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Represent in matrix form: AX = B
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

B_eq = np.array([-1, -3, 9])


X = np.dot(np.linalg.inv(A_eq), B_eq)
print("\nSolution using inverse method:")
print("x, y, z =", X)


X_solve = np.linalg.solve(A_eq, B_eq)
print("\nSolution using np.linalg.solve (for verification):")
print("x, y, z =", X_solve)
```

```python
import numpy as np
import time

# 1. Element-wise Addition
size = 1_000_000

# Using Python lists
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
list_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python list addition time: {:.5f} seconds".format(end - start))

# Using NumPy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy array addition time: {:.5f} seconds".format(end - start))
```
